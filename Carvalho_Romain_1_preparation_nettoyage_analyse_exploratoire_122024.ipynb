```python
# Création d'une liste pour ne garder que celles utiles
liste_elements_alimentation = (
    'Production', 'Importations - Quantité', 'Variation de stock',
    'Exportations - Quantité', 'Disponibilité intérieure', 'Nourriture',
    'Disponibilité alimentaire en quantité (kg/personne/an)',
    'Disponibilité alimentaire (Kcal/personne/jour)',
    'Disponibilité de protéines en quantité (g/personne/jour)',)

# Filtrer les données pour ne garder que les éléments présents dans la liste
df_dispo_alim = df_dispo_alim[df_dispo_alim['Élément'].isin(liste_elements_alimentation)]

# Afficher les premières lignes des données filtrées
print(df_dispo_alim.head())
```

```python
# Pour l'étude de marche nous ne garderons que Viand de Volailles
df_dispo_alim = df_dispo_alim.loc[df_dispo_alim['Produit'] == 'Viande de Volailles', :]
```

```python
df_dispo_alim
```

```python
df_dispo_alim_pivot = pd.pivot(df_dispo_alim,
                                     index=['Zone'],
                                     columns=['Élément'],
                                     values=['Valeur'])

display(df_dispo_alim_pivot.head())
```

```python
# Compter les valeurs manquantes (NaN) par colonne
print(df_dispo_alim_pivot.isnull().sum())
# Afficher les lignes avec des valeurs manquantes
print(df_dispo_alim_pivot[df_dispo_alim_pivot.isnull().any(axis=1)])
```

```python
# Tentons de calculer les valeurs manquantes afin de biaiser les calculs au minimum
df_dispo_alim_pivot[('Valeur', 'Exportations - Quantité')] = df_dispo_alim_pivot.apply(
    lambda x: ((x[('Valeur', 'Production')] + x[('Valeur', 'Importations - Quantité')] - 
               x[('Valeur', 'Disponibilité alimentaire en quantité (kg/personne/an)')] + 
               x[('Valeur', 'Variation de stock')]) 
              - x[('Valeur', 'Disponibilité intérieure')])
    if pd.isnull(x[('Valeur', 'Exportations - Quantité')]) else x[('Valeur', 'Exportations - Quantité')], axis=1)

# Afficher les premières lignes après calcul
print(df_dispo_alim_pivot.head())
```

```python
# Compter les valeurs manquantes (NaN) par colonne
print(df_dispo_alim_pivot.isnull().sum())
```

```python
# Remplacer les NaN par 0 dans le DataFrame pivoté
df_dispo_alim_pivot = df_dispo_alim_pivot.fillna(0)
# Compter les valeurs manquantes (NaN) par colonne
print(df_dispo_alim_pivot.isnull().sum())
```

```python
# Tracer une boîte à moustaches (boxplot) pour 'Exportations - Quantité'
df_dispo_alim_pivot.boxplot(column=[('Valeur', 'Exportations - Quantité')])

# Afficher le graphique
plt.title("Boxplot des Exportations - Quantité")
plt.ylabel("Quantité")
plt.show()
```

```python
# Vérification du pays avec une très grande quantité d'exportations
print(df_dispo_alim_pivot.sort_values(by=('Valeur','Exportations - Quantité'), ascending=False))
```

```python
# Filtrer pour ne garder que les informations relatives au PIB et à la Stabilité Politique
df_pib_stab = df_pib_stab[df_pib_stab['Produit'].isin(
    ['PIB par habitant, ($ PPA internationaux constants de 2017)',
     'Stabilité politique et absence de violence/terrorisme (indice)'])]

# Pivot pour créer des colonnes séparées pour le PIB et la Stabilité Politique
df_pib_stab = df_pib_stab.pivot(index='Zone', columns='Produit', values='Valeur')

# Renommer les colonnes pour plus de clarté
df_pib_stab.columns = ['PIB', 'Stabilité Politique']

# Afficher les premières lignes du DataFrame nettoyé
print(df_pib_stab.head())
```

```python
df_dispo_alim_pivot.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_dispo_alim_pivot.columns]

# Afficher les premières colonnes pour vérifier
print(df_dispo_alim_pivot.head())
```

```python
df_dispo_alim_stab = pd.merge(df_dispo_alim_pivot, df_pib_stab, on='Zone', how='outer')
# Afficher les premières lignes du DataFrame après la jointure
df_dispo_alim_stab
```

```python
# Vérifier les valeurs manquantes
print(df_dispo_alim_stab.isnull().sum())
```

```python
df_dispo_alim_stab
```

```python
# Afficher toutes les lignes contenant des valeurs manquantes
rows_with_missing_values = df_dispo_alim_stab[df_dispo_alim_stab.isnull().any(axis=1)]

# Afficher les premières lignes avec des valeurs manquantes
print("Lignes avec des valeurs manquantes :")
print(rows_with_missing_values)
```

```python
# Réinitialiser l'index pour remettre 'Zone' comme colonne
df_dispo_alim_stab_reset = df_dispo_alim_stab.reset_index()

# Vérifier que 'Zone' est bien une colonne
print(df_dispo_alim_stab_reset.head())
```

```python
# Remplacer les variations dans les noms de la Chine pour les standardiser
df_dispo_alim_stab_reset['Zone'] = df_dispo_alim_stab_reset['Zone'].replace({
    'Chine, Taiwan Province de': 'Chine',
    'Chine, continentale': 'Chine'
})

# Vérifier les occurrences après correction
print(df_dispo_alim_stab_reset[df_dispo_alim_stab_reset['Zone'] == 'Chine'])
```

```python
# Regrouper par 'Zone' en faisant la somme des valeurs numériques pour chaque pays
df_dispo_alim_stab_grouped = df_dispo_alim_stab_reset.groupby('Zone').sum().reset_index()

# Vérifier que 'Chine' est maintenant correctement regroupée
print(df_dispo_alim_stab_grouped[df_dispo_alim_stab_grouped['Zone'] == 'Chine'])
```

```python
df_merged = df_dispo_alim_stab_grouped
```

```python
# Vérifier le nombre de valeurs manquantes pour chaque colonne
missing_values = df_merged.isnull().sum()

# Afficher les colonnes avec des valeurs manquantes
print("Valeurs manquantes par colonne :")
print(missing_values[missing_values > 0])

# Afficher toutes les lignes contenant des valeurs manquantes
rows_with_missing_values = df_merged[df_merged.isnull().any(axis=1)]

# Afficher les premières lignes contenant des valeurs manquantes
print("Lignes avec des valeurs manquantes :")
print(rows_with_missing_values)

# Vérifier s'il y a des doublons dans le DataFrame
duplicate_rows = df_merged[df_merged.duplicated()]

# Afficher les doublons s'ils existent
print("Doublons trouvés :")
print(duplicate_rows)
```

```python
# Fusionner les données de population avec le DataFrame df_merged
df_final_merged = pd.merge(df_merged, df_population_2017[['Zone', 'Valeur']], on='Zone', how='left')

# Renommer la colonne 'Valeur' en 'Population' pour plus de clarté
df_final_merged.rename(columns={'Valeur': 'Population'}, inplace=True)

# Vérifier les premières lignes après la fusion
print(df_final_merged.head())
```

```python
# Vérifier les valeurs manquantes dans le DataFrame fusionné
missing_values_population = df_final_merged.isnull().sum()

print("Valeurs manquantes après la fusion :")
print(missing_values_population[missing_values_population > 0])


### Check Valeur null
```

```python
# Multiplier la colonne 'Population' par 100 pour corriger les valeurs
df_final_merged['Population'] = df_final_merged['Population'] * 100

# Vérifier les premières lignes après la correction
df_final_merged
```

```python
# Filtrer et conserver uniquement les pays ayant une population de 300 000 ou plus
df_final = df_final_merged[df_final_merged['Population'] >= 300000]

# Vérifier les premières lignes après suppression des pays
print(df_final.head())



## Fonction 60%> de la pop mondiale pour analyse 
```

```python
# Calculer la somme de la population dans le DataFrame final
total_population = df_final['Population'].sum()

# Afficher la somme totale de la population
print("Somme totale de la population : ", total_population)
```

```python
# Remettre 'Zone' comme index
df_final.set_index('Zone', inplace=True)

# Vérifier les premières lignes pour confirmer que 'Zone' est bien l'index
df_final
```

```python
# Renommer les colonnes du DataFrame
df_final.rename(columns={
    'Valeur_Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité alimentaire (kg/personne/an)',
    'Valeur_Disponibilité de protéines en quantité (g/personne/jour)': 'Disponibilité de protéines (g/personne/jour)',
    'Valeur_Disponibilité intérieure': 'Disponibilité intérieure',
    'Valeur_Exportations - Quantité': 'Exportations',
    'Valeur_Importations - Quantité': 'Importations',
    'Valeur_Production': 'Production',## +
    'Valeur_Nourriture': 'Nourriture',## +
    'Valeur_Variation de stock': 'Variation de stock',## --
    'PIB': 'PIB par habitant',
    'Stabilité Politique': 'Stabilité politique',
    'Population': 'Population'
}, inplace=True)

# Vérifier les premières lignes après renommage
print(df_final.head())
```

```python
# Ne garder que les colonnes spécifiées dans le DataFrame avec les bons noms de colonnes
df_final = df_final[[
    'Disponibilité alimentaire (kg/personne/an)', 
    'Disponibilité de protéines (g/personne/jour)',
    'Disponibilité intérieure',
    'Exportations', 
    'Importations',
    'PIB par habitant', 
    'Stabilité politique', 
    'Population'
]]

df_final
```

```python
# Sélectionner les colonnes importantes (par exemple, PIB, Population, Disponibilité alimentaire)
colonnes_importantes = ['PIB par habitant', 'Population', 'Disponibilité alimentaire (kg/personne/an)']

# Créer un boxplot pour chaque colonne
plt.figure(figsize=(15, 10))

for i, colonne in enumerate(colonnes_importantes, 1):
    plt.subplot(1, len(colonnes_importantes), i)
    sns.boxplot(data=df_final, y=colonne)
    plt.title(f'Boxplot de {colonne}')

plt.tight_layout()
plt.show()
```
